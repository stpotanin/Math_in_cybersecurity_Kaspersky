```python
caesar = lambda x, key: chr((ord(x) - ord(key)) % 32 + ord('А')) if ord('А') <= ord(x) <= ord('Я') else x

message, word = 'ШШЯССАБРШЭЭЧШЫ', 'ОКТЯБРЬ'
key = ((len(message) // len(word) + 1) * word)[:len(message)] # ключ нужной длины

print(''.join(caesar(x, i) for x, i in zip(message, key))) # расшифровка
```

```python
# Василий Гурьев https://stepik.org/lesson/288967/step/8?discussion=3008069&thread=solutions
n = 32
abc = [chr(ord('А') + i) for i in range(n)]
key = 'ОКТЯБРЬ'
code = 'ШШЯССАБРШЭЭЧШЫ'
print(''.join([abc[(abc.index(c) - abc.index(key[i % len(key)])) % n] for i, c in enumerate(code)]))
```
