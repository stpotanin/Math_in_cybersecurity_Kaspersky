```python
# @title https://stepik.org/lesson/289979/step/8 Найти все первообразные корни по модулю 6
gcd = lambda a, b: abs(a) if b == 0 else gcd(b, a % b)

phi_simple = lambda n: sum([gcd(i, n) == 1 for i in range(1, n)])

def primitive_root_modulo_n(n):
    roots = []
    phi = phi_simple(n)
    print('phi =', phi)
    for g in range(2, n):
        no_one = True
        for k in range(1, phi):
            if g**k % n == 1:
                no_one = False
                break
        if (g**phi % n == 1) and no_one:
                roots.append(g)
    return roots if n>1 else [1]

primitive_root_modulo_n(4)
```
