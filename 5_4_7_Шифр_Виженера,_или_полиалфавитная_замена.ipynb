```python
caesar = lambda x, key: chr((ord(x) + ord(key)) % 32 + ord('А')) if ord('А') <= ord(x) <= ord('Я') else x

message, word = 'И ДЫМ ОТЕЧЕСТВА НАМ СЛАДОК И ПРИЯТЕН', 'ПОЛНАЯ ПОБЕДА'
m, w = message.replace(' ', ''), word.replace(' ', '') # удаляем пробелы
key = ((len(m) // len(w) + 1) * w)[:len(m)] # ключ нужной длины

cipher = [caesar(x, i) for x, i in zip(m, key)] # шифр списком без пробелов

print(''.join(cipher.pop(0) if x!=' ' else ' ' for x in message)) # добавляем пробелы
```

```python
# Василий Гурьев https://stepik.org/lesson/288967/step/7?discussion=3008042&thread=solutions
n = 32
abc = [chr(ord('А') + i) for i in range(n)]
key = 'ПОЛНАЯ ПОБЕДА'.replace(' ', '')
text = 'И ДЫМ ОТЕЧЕСТВА НАМ СЛАДОК И ПРИЯТЕН'
spaces = 0
code = ''
for i, c in enumerate(text):
    if c == ' ':
        spaces += 1
        code += c
    else:
        code += abc[(abc.index(c) + abc.index(key[(i - spaces) % len(key)])) % n]
print(code)
```
