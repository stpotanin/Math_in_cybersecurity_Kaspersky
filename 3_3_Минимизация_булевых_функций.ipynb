```python
# @title https://stepik.org/lesson/289980/step/3
# Синтаксис
from sympy import symbols
from sympy.logic.boolalg import And, Or, Not, simplify_logic
from sympy.abc import x, y, z

# Создаем логическое выражение
expr = And(Or(x, y), Not(Or(x, z)))
print(expr)

# Упрощаем
simplified = simplify_logic(expr)
print(simplified)  # x | y & ~z
```

```python
expr = Or(And(x, y), And(x, Not(y)))
simplify_logic(expr)  # x
```

```python
from sympy import symbols
from sympy.logic.boolalg import Implies

x, y = symbols('x y')

# Функция Implies(P, Q) = P → Q
# Implies(x, y) или x >> y
# imp = Implies(x, y)
imp = x >> y
simplify_logic(imp)
```

```python
from sympy import symbols
from sympy.logic.boolalg import And, Or, Not, Implies, simplify_logic

x1, x2, x3 = symbols('x1 x2 x3')

expr = And(Not((Not(x1) >> Not(x2)) >> And(x2, x3) >> And(Not(x1), x3)), x3)
simplify_logic(expr)
```
