```python
coincidence_index = lambda t: sum(f * (f - 1) for f in [t.replace(' ', '').count(i) \
    for i in set(t.replace(' ', ''))]) / len(t.replace(' ', '')) / (len(t.replace(' ', '')) - 1)

text = 'ВАРИС ВАРИТ ВАТУ'

round(coincidence_index(text), 4)
```
