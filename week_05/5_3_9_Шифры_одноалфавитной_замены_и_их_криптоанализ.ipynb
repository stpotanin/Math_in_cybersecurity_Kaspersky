```python
cyrillic = ''.join([chr(x) for x in list(range(ord('А'), ord('Я')+1))])
cipher = 'ГЗЩПБОФЪШЭАУЯЙЦДЮКНЕЖМТВСРЧЛЫИЬХ'
print(set(cyrillic) == set(cipher)) # проверка на однозначность

m = 'ЖИЛ-БЫЛ СТАРИЧОК ИЗ ГОНКОНГА, ТАНЦЕВАВШИЙ ПОД МУЗЫКУ ГОНГА. НО ЕМУ ЗАЯВИЛИ: ПРЕКРАТИ ЭТО, ИЛИ УБИРАЙСЯ СОВСЕМ ИЗ ГОНКОНГА!'

# ТЕЛО ШИФРАТОРА
# Наполняем словарь для замены
# d = {cyrillic[i]: cipher[i] for i in range(len(cyrillic))}
d = dict(zip(cyrillic, cipher))

# Замена по словарю
print(''.join(d.get(x, x) for x in m))
```
