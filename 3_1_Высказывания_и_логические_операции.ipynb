```python
# @title https://stepik.org/lesson/288953/step/11

[print(int(not x1 or (not (x2 and x3)) or not x1 and x3), end='') for x in range(8) for x1, x2, x3 in [(x // 4 % 2, x // 2 % 2, x % 2)]]; print()
```

```python
# @title https://stepik.org/lesson/288953/step/13

[print(int(not(x1 and x2 or not x2)), end='') for x in range(4) for x1, x2 in [(x // 2 % 2, x % 2)]]; print()
```
