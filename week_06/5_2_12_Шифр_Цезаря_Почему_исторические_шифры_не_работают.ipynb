```python
# @title https://stepik.org/lesson/288959/step/11
decipher = lambda x, key: chr((ord(x) - key) % 32 + ord('А')) if ord('А') <= ord(x) <= ord('Я') else x

message = 'КЗТТВ ЙВ КТЪЦЭВ УХР ШМСЧМЩ ЙЗУ ФМ ЛХШЩЗФМЩШЖ'

print('\n'.join([f"{ ''.join(decipher(x, i) for x in message)} key = {(i-ord('А'))%32}" for i in range(31)]))

# Пришлось модифицировать подсчёт ключа, так как по факту мы начинаем крутиться от ord('А'), а эта точка не равна нулю по модулю 32.
```
