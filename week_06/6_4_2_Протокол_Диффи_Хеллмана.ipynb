**Важно**: первообразный корень по модулю n существует тогда и только тогда, когда n — либо степень нечётного простого, либо удвоенная степень простого, а также в случаях n = 1, n = 2, n = 4


```python
# Модификация функции для поиска всех корней:
'''Параметры функции:
p - модуль (простое число)
phi - значение функции Эйлера от p (обычно p-1 для простого p)
fact - список простых множителей числа phi'''
def find_all_primitive_roots(p, phi, fact):
    roots = []
    for res in range(2, p):
        ok = True
        for f in fact:
            if pow(res, phi // f, p) == 1:
                ok = False
                break
        if ok:
            roots.append(res)
    return roots

# Пример для p = 5
p = 5
phi = p - 1  # 4
fact = [2]  # простые множители числа 4
find_all_primitive_roots(p, phi, fact)
```
