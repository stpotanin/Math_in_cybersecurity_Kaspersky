$c \equiv f(m) \equiv m^{e} \pmod{n}$ - легко вычисляется, **m** и **c** - текст и шифртекст, (**e**, **n**) - открытый ключ.

$m \equiv f^{-1}(c) \equiv c^{d} \pmod{n}$ - легко вычисляется при известном закрытом ключе (**d**, **n**).

$d \equiv e^{-1} \pmod{\varphi(n)}$ - сложно без факторизации, требует знания $φ(n)$, а оно зависит от факторизации $n=pq$.

Поскольку открытый ключ (e, n) = (27, 55), легко найти $\varphi(n)$.

Разложим 55 на простые числа: $55 = 11 * 5$

По свойству мультиплпкативности функции Эйлера:



$\varphi(n) = \varphi(pq) = \varphi(p)\varphi(q) = (p-1)(q-1) = (11-1)(5-1) = 40$

Теперь легко найти недостающую часть $d$ закрытого ключа:

$d \equiv e^{-1} \pmod{\varphi(n)} = 27^{-1}\pmod{40} = 3$

Код для получения обратного элемента по модулю:

```python
# Расширенный НОД
# возвращает не только gcd, но и коэффициенты x, y из ax+by=gcd(a,b).
def egcd(a, b):
    if a == 0:
        return b, 0, 1
    g, y, x = egcd(b % a, a)
    return g, x - (b // a) * y, y

# Обратный элемент по умножению
def mod_inverse(a, m):
    g, x, y = egcd(a, m)
    if g != 1:
        raise ValueError("Обратный элемент не существует")
    return x % m

d = mod_inverse(27, 40)
print('d =', d)
```

Находим буквы нашего последнего слова:

```python
decoder = lambda c, d, n: c**d % n

ciphertext = [28, 23, 25, 26]
d, n = 3, 55 # закрытый ключ

text = ''.join([chr(decoder(c, d, n) + ord('А')) for c in ciphertext])
print(text)

# # Русский алфавит без Ё
# cyrillic = [chr(x) for x in list(range(ord('А'), ord('Е')+1)) + [ord('Ё')] + list(range(ord('Ж'), ord('Я')+1))]
# print(''.join(cyrillic))
```
