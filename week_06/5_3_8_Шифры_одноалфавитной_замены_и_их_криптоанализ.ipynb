```python
# @title Сортируем буквы в сообщении по частотам
m = "ФЙИ-ЭЕИ ЦСИЧМСО М ШДТЫСЮБШДС, ЩС ЦЙТЫЙМЖЙЬ ЖИПГА УЧБШДЙ. ЧЩ М ЩСЬ ЩСМЛЩШЦШЬ ЛШМШЮЙМШИ ЦШЬ Й М ЩСЬ ФС УАИПИ М ШДТЫСЮБШДС."

alphabet = sorted(
    set(x for x in m if ord('А') <= ord(x) <= ord('Я')),
    key=lambda x: m.count(x),
    reverse=True
)

print(''.join(alphabet))
```

```python
# @title Замены по часте
# Классическая частотность в русском языке без Ё
# frequency = 'ОЕАИНТСЛВРКМДПЫУБЯЬГЗЧЙЖХШЮЦЭЩФЪ' # классика 1
# frequency = 'ОЕАИНТСРВЛКМДПЫУЯЬГЗБЧЙХЖШЮЦЭЩФЪ' # классика 2
alphabet_ = 'ШСМЙИЩДЬЦТБЧЮЫФУАЖЛПОЕГЭ' # просто записываем, чтобы легче двигать
frequency = 'АЕВИЛНМЙЧСДОРТЖГУШЗЯКЫПБХЬЮЦЭЩФЪ' # здесь двигаем буквы под ЖИЛ-БЫЛ
'''ВЫШЕ ГЛАВНАЯ РАБОЧАЯ СТРОКА'''

# 1) двигаем буквы под ФЙИ-ЭЕИ ... М == ЖИЛ-БЫЛ ... В
# 2) двигаем буквы под ЩСЬ == НЕМ (ошибка, надо НЕЙ)
# 3) двигаем буквы под ЦСИЧМСО == ЧЕЛОВЕК
# 4) двигаем буквы под ШДТЫСЮБШДС == АМСТЕРДАМЕ
# 5) двигаем буквы под ЦЙТЫЙМЖЙЬ == ЧИСТИВШИЙ
# 6) двигаем буквы под ЖИПГА УЧБШДЙ == ШЛЯПУ ГОДАМИ

# ТЕЛО ДЕШИФРАТОРА
# Наполняем словарь для замены
d = {alphabet[i]: frequency[i] for i in range(len(alphabet))}

# Замена по словарю
print(''.join(d.get(x, x) for x in m))
```
