```python
caesar = lambda x, key: chr((ord(x) + ord(key)) % 32 + ord('А')) if ord('А') <= ord(x) <= ord('Я') else x

message = 'ЭКЗИСТЕНЦИАЛИЗМ'
w = 'ОМГ'
cipher = (len(message) // len(w) + 1) * w
cipher = cipher[:len(message)]

print(''.join(caesar(x, i) for x, i in zip(message, cipher)))
```
