```python
import hashlib

sha256_hex = lambda text: hashlib.sha256(text.encode()).hexdigest()

text1 = "You shall not pass!"
text2 = "You shall not past!"

hash1 = sha256_hex(text1)
hash2 = sha256_hex(text2)

sum(a == b for a, b in zip(hash1, hash2))

```
