```python
caesar = lambda x, key: chr((ord(x) - ord(key)) % 32 + ord('–Р')) if ord('–Р') <= ord(x) <= ord('–ѓ') else x

message, word = '–§–†–Ч–Ю–†–†–Ґ–ѓ–Ђ–†–Ч–У–Ґ–≠–Х–•–Ъ–§–ѓ–С–Ѓ–§–Ь–Ђ–Ю–Ь–Ю–ѓ–ђ', '–†–Р–Ч–Ф–†–Р–Ъ–Ю–Э–Ф–Т–Р–Ф–†–Р–Ъ–Ю–Э–Р–Ґ–†–Ш–Ф–†–Р–Ъ–Ю–Э–Р'
key = ((len(message) // len(word) + 1) * word)[:len(message)] # –Ї–ї—О—З –љ—Г–ґ–љ–Њ–є –і–ї–Є–љ—Л

print(''.join(caesar(x, i) for x, i in zip(message, key))) # —А–∞—Б—И–Є—Д—А–Њ–≤–Ї–∞
```
