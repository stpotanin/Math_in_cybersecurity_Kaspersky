```python
# @title https://stepik.org/lesson/288959/step/9
cipher = lambda letter, key: chr((ord(letter) + key) % 32 + ord('А')) if ord('А') <= ord(letter) <= ord('Я') else letter

message = 'НАСТУПАЙТЕ НОЧЬЮ'
key = 19 # потому что отсчитывали от ord('А'), а не от нуля.

print(''.join(cipher(x, key) for x in message))
```
